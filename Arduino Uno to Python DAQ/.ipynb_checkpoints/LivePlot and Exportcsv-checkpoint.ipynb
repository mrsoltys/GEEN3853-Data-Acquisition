Python DAQ
==========

Install Packages.

Note: If you get an error "No module found" you may need to install packages in your python environment. I like to manage all of this through anaconda navigator. 

```python
%matplotlib widget              
# Necessary for live plots
import time
import serial                   # from pyserial (Needs installed) for reading data from Arduino over serial
import pandas as pd             # Package for managing datasets
from collections import deque   # double-ended queue, needed for quickly adding and removing data in live plot.
import matplotlib.pyplot as plt # Package for plotting
from matplotlib.animation import FuncAnimation # For live plot animation
import threading                # Allows us to do two things at once (import data _and_ plot)
```

Connect to Arduino. 
-------------------
The first script will list Serial Pots. You can then adjust the code in the second script to connect to the correct Serial Port

```python
import serial.tools.list_ports

ports = serial.tools.list_ports.comports()

if not ports:
    print("No serial ports found.")
else:
    print("Available serial ports:")
    for p in ports:
        print(f"  {p.device}  —  {p.description}")
```

Change `PORT = ""` to match your serial port name

```python
PORT = "/dev/cu.usbmodem1201"   # mac example (or "COM5" on Windows)
BAUD = 115200

ser = serial.Serial(PORT, BAUD, timeout=1) #Connect to serial device
time.sleep(2)  # allow Arduino reset
```

**Optional:** Run this cell to read a single line from arduino

```python
line = ser.readline().decode(errors="ignore")
print(line)
```

Begin Reading Data
------------------
Running this cell will begin reading data... untill we change the `stop_flag` to `TRUE`

```python
# Buffers for plotting
N = 500
tbuf = deque(maxlen=N)
vbuf = deque(maxlen=N)

# Buffer for CSV logging (grows for entire run)
log_rows = []

stop_flag = False
t0_pc = time.time()

def parse_line(s: str):
    s = s.strip()
    if not s or s.lower().startswith("us"):
        return None
    parts = s.split(",")
    if len(parts) != 2:
        return None
    try:
        us = int(parts[0].strip())
        raw = int(parts[1].strip())
       # volts = float(parts[2].strip())
       # return ms, raw, volts
        return us, raw
    except ValueError:
        return None

def reader():
    global stop_flag
    ser.reset_input_buffer()
    while not stop_flag:
        line = ser.readline().decode("utf-8", errors="ignore")
        parsed = parse_line(line)
        if parsed:
            us, raw = parsed

            # Plot buffers
            tbuf.append(us / 1000000.0)
            vbuf.append(raw * 5/ 1023)

            # Log buffer (full run)
            log_rows.append({
                "pc_time_s": time.time() - t0_pc,
                "arduino_us": us,
                "raw": raw
            })

thread = threading.Thread(target=reader, daemon=True)
thread.start()

print("Reader thread started.")
```

Live plot of data
-----------------

Data is already being recorded, but running this cell will start a live plot of the data output. 

```python
fig, ax = plt.subplots()
(line,) = ax.plot([], [])
ax.set_xlabel("Arduino time (s)")
ax.set_ylabel("Volts")
ax.grid(True)
ax.set_title("Live DAQ (Arduino → Python)")

def init():
    line.set_data([], [])
    return (line,)

def update(frame):
    if len(tbuf) < 2:
        return (line,)
    x = list(tbuf)
    y = list(vbuf)
    line.set_data(x, y)
    ax.relim()
    ax.autoscale_view()
    return (line,)

ani = FuncAnimation(
    fig, update,
    init_func=init,
    interval=50,
    blit=False,
    cache_frame_data=False
)

plt.show()
```

Stop recording data, and write file to CSV
------------------------------------------

Running this cell will stop the recording and write to CSV. 

You may wish to change the file name `"daq_data.csv"` so you don't overwrite other data files. 

```python
# 1) Stop the reader thread
stop_flag = True
thread.join(timeout=1)

# 2) Stop the animation timer
ani.event_source.stop()

# 3) Close the serial port
ser.close()

# 4) Close the figure (important in notebooks)
import matplotlib.pyplot as plt
plt.close(fig)

# 5) Write the file to CSV
df = pd.DataFrame(log_rows)
df.to_csv("daq_data.csv", index=False)

print(f"Saved {len(df)} samples to daq_data.csv")
df.head()
```
